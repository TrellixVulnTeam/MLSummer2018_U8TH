# Logistic regression

Logistic regression is an algorithm for binary classification. Logistic regression is **NOT** a linear model for continuous data. It is available in scikit-learn as

```python
from sklearn.linear_model import LogisticRegression
```

Take a look at the [documentation](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

## Logit Function

Given a probability $p$, the function $\text{logit}(p)$ function gives the **logarithm of the odds**

$$\text{logit}(p) = \log\left(\frac{p}{1-p}\right) $$

Note that the range is $(-\infty,\infty)$, just like the possible results of dot products. So we make the stipulation

$$\text{logit}(p)=\mathbf{w}^T\mathbf{x}$$

where $p$ is the probability of the input $\mathbf{x}$ being in class $y=1$.

## Sigmoid (Mirror Image of Fermi-Dirac)

Inverting the $\text{logit}$ function yields

$$\text{logit}(p)=\log{\left(\frac{p}{1-p}\right)} = \mathbf{w}^T\mathbf{x}=z \implies p = \frac{1}{1+\exp{-z}}$$

This is called a "sigmoid" function, although physicists recognize it as being rather similar to the Fermi-Dirac distribution.

We can implement this in a `Pereptron`. Recall that one of our earlier `Perceptron` models used a Heaviside function for classification. We broaden it with the sigmoid

$$ \phi(z) = \frac{1}{1+\exp{-z}}$$

We interpret this as the **probability a sample belongs to class y=1**.

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np


def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
# plt.savefig('./figures/sigmoid.png', dpi=300)
plt.show()
````

An input $\mathbf{x}$ is then in class 1 if $\phi(z)>0.5$ and in the other class otherwise. That is,

$$ \hat y = \left\{\begin{array}{} 1 & \phi(z) \geq 0.5 \\
                                   0 & \phi(z) < 0.5 \end{array}\right.$$

## The Logistic cost function for Binary Classification

To understand the logistic cost function, we first need to define the *likelihood*. That is, the probability that we correctly label all inputs $\mathbf{x}^(i)$ as $\mathbf{y}^{(i)}$ given the weights we chose $\mathbf{w}$.

$$\textbf{Likelihood: } L(\mathbf{w})= P(\mathbf{y}|\mathbf{x};\mathbf{w}) $$

Assuming the inputs are independent of one another, this is a product of each individual probability

$$L(\mathbf{w}) = \prod_{i} P(y^{(i)}|x^{(i)};\mathbf{w}) = \prod_{i}\phi(z)^{y^{(i)}}(1-\phi(z))^{1-y^{(i)}}$$

To make derivatives easier and scale the probabilities away from ridiculously small values, we take the log of the likelihood $\ell(\mathbf{w})=\log{L(\mathbf{w})}$ and maximize that. However, we want to think of it like a cost function, so we **minimize the negative log-likelihood**.

$$J(\mathbf{w}) = -\ell(\mathbf{w})=-\log(L(\mathbf{w}))$$
$$J(\mathbf{w}) = -\sum_i y^{(i)}\log{(\phi(z^{(i)}))} + (1-y^{(i)})\log{(1-\phi(z^{(i)}))}$$

where $z^{(i)}=\mathbf{w}^T\mathbf{x}^{(i)}$, and $\phi(z)$ is the sigmoid function.

<br>

<details>
    <summary>Typo warning</summary>
*The above equation has a typo on Raschka's notes. It omits the mult. factor $y^{(i)}$*
</details>

As always, we updates weights $\mathbf{w}$ using $\mathbf{w}\to \mathbf{w}-\eta\nabla J(\mathbf{w})$.

````python
def cost_1(z):
    return - np.log(sigmoid(z))


def cost_0(z):
    return - np.log(1 - sigmoid(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)

c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='J(w) if y=1')

c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='J(w) if y=0')

plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.xlabel('$\phi$(z)')
plt.ylabel('J(w)')
plt.legend(loc='best')
plt.tight_layout()
# plt.savefig('./figures/log_cost.png', dpi=300)
plt.show()
````

## Training a logistic regression model with scikit-learn

### 1. Arrange the data into features matrix and target vector

Reload the iris data.

````python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
````

### 2. Split data into train and test data

Split once again into training and classifying data.

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
````

````python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
````

### 3. Choose a class of model by importing the appropriate estimator from scikit-learn

````python
from sklearn.linear_model import LogisticRegression
````

### 4. Choose model hyperparameters by instantiating this class with desired values

````python
lr = LogisticRegression(C=1000.0, random_state=0)
````

### 5. Fit the model to your data by calling the ``fit()`` method of the model instance.

````python
lr.fit(X_train_std, y_train)
````

### 6. Apply the model to new data

````python
y_pred = lr.predict(X_test_std)
````

### 7. Assess performance of the model

````python
from sklearn.metrics import accuracy_score

print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
````

Redefining the `plot_decision_region` function from chapter 2:

### Visualizing the Solution

````python
from PlotDecision import plot_decision_regions

X_combined_std = sc.transform(X)
y_combined = y

plot_decision_regions(X_combined_std, y_combined,
                      classifier=lr, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
# plt.savefig('./figures/logistic_regression.png', dpi=300)
plt.show()
````

### Tackling overfitting via regularization

````python
Image(filename='./images/03_06.png', width=700) 
````

````python
weights, params = [], []
for c in np.arange(-5., 5.):
    lr = LogisticRegression(C=10.**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)

weights = np.array(weights)
plt.plot(params, weights[:, 0],
         label='petal length')
plt.plot(params, weights[:, 1], linestyle='--',
         label='petal width')
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
# plt.savefig('./figures/regression_path.png', dpi=300)
plt.show()
````
